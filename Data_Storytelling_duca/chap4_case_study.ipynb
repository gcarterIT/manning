```python
## Listing 4.7 Generating the stacked bar chart

# Import the required libraries
import pandas as pd
import altair as alt

# Load '../data/competitions.csv' as pandas dataframe
df = pd.read_csv('E:\\docs\\book_reviews\\manning\\competitions.csv')

# Select only the columns Training Type, Record Time (Seconds), Our Best Time
df = df[['Training Type', 'Record Time (Seconds)', 'Our Best Time']]

# Use melt to transform the dataframe into a long format, with the parameter id_vars set to 'Training Type'
df = pd.melt(df, id_vars=['Training Type'])

# Draw a bar chart named chart in Altair with:
# * The Training Type column as the x-axis
# * The value column as the y-axis
# * The variable column as the color

chart = alt.Chart(df).mark_bar().encode(
    x='Training Type',
    y='value',
    color='variable'
)

# Save the chart as 'competitions.html'
chart.save('E:\\docs\\book_reviews\\manning\\competitions_4_7.html')
```

```python
# Figure 4.19 percentage improvement

import pandas as pd
import altair as alt

# Load 'E:\\docs\\book_reviews\\manning\\competitions.csv' as pandas dataframe
df = pd.read_csv('E:\\docs\\book_reviews\\manning\\competitions.csv')

# Select only the columns Training Type, Record Time (Seconds), Our Best Time
df = df[['Training Type', 'Record Time (Seconds)', 'Our Best Time']]

# Calculate the percentage difference between columns Our Best Time and Record Time (Seconds)  and store it into a new column called Percentage Difference
df['Percentage Difference'] = (df['Our Best Time'] - df['Record Time (Seconds)']) / df['Record Time (Seconds)'] * 100

df['Percentage Improvement'] = 100 - (df['Our Best Time'] - df['Record Time (Seconds)']) / df['Record Time (Seconds)'] * 100

chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('Training Type', sort='-y'),
    y=alt.Y('Percentage Improvement', scale=alt.Scale(domain=[0, 100]))
).properties(
    width=300#,
    #height=200,
    #title='Percentage Difference between Our Best Time and Record Time (Seconds)',
    #color='Training Type'
)

# chart

# Save the chart 
chart.save('E:\\docs\\book_reviews\\manning\\competitions_fig_4_19.html')
```

```python
# Listing 4.12 Highlighting the sports that perform better

# Add the color encoding. Set the color to:
    # - #003049 if the Percentage Improvement is greater than 50, 
    # - lightgray otherwise
chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('Training Type', sort='-y'),
    y=alt.Y('Percentage Improvement', scale=alt.Scale(domain=[0, 100])),
    color=alt.condition(
        alt.datum['Percentage Improvement'] > 50,
        alt.value('#003049'),
        alt.value('lightgray')
    )
).properties(
    width=300#,
    #height=200,
    #title='Percentage Difference between Our Best Time and Record Time (Seconds)',
    #color='Training Type'
)

chart
```
