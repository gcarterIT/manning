```python
# Import the required libraries
# Load the file '/path/to/data.csv' into a pandas DataFrame
# Print the first 5 rows of the DataFrame
import pandas as pd
df = pd.read_csv('E:\\temp\\data3.csv')
print(df.head())


# Draw a bar chart named bar in Altair with:
# * The category column as the x-axis 
# * The value column as the y-axis
# * The country column as the color.
# * The chart should have a title.
import altair as alt
bar = alt.Chart(df).mark_bar().encode(
    x='category',
    y='value',
    color='country',
).properties(
    title='Bar Chart'
)
bar

```

```python
# * The country column as the color. Set the color to:
#   - 'red' for 'IT'
#   - 'green' for 'FR'
#   - 'blue' for 'DE'
#   - 'orange' for 'UK' 
#   - 'purple' for 'US'
bar = alt.Chart(df).mark_bar().encode(
    x='category',
    y='value',
#    color=alt.Color('country', scale=alt.Scale(domain=['IT', 'FR', 'DE', 'UK', 'US'], range=['red', 'green', 'blue', 'orange', 'purple']))
    color=alt.condition(
            alt.datum.country == 'Italy',
            alt.value('red'),
            alt.value('green'))
).properties(
    title='Bar Chart'
)
bar
```

```python
df = pd.read_csv('E:\\temp\\data.csv')

line = alt.Chart(df).mark_line().encode(
    x='X',
    y='Y'
)

line
```

```python
import altair as alt

# Build a compound chart named chart with the line and bar charts aligned vertically
chart = alt.vconcat(bar, line)
chart
```

```python
# Make the chart interactive
chart.interactive()
  
```
