```python
import numpy as np
import pandas as pd

import camelot

from IPython.display import display
```

Let's create a reusable function to extract and print all the tables in a PDF

```python
def extract_tables(pdf_path, pages="all", print_tables=True):
    """
    Extract all tables found in a PDF.
    
    `pdf_path`: file path pointing to the PDF
    `pages`: the page number(s) to read
    `print_tables`: whether to also print out all the tables that are found (default: True)
    
    returns: a list of pandas DataFrames
    """
    
    print(f"Reading {pdf_path}")
    
    tables = camelot.read_pdf(pdf_path, pages=pages, flavor="stream")
    
    print(f"{tables.n} tables found.")
    
    if tables.n > 0:
        dataframes = [table.df for table in tables]
        
        if print_tables:
            for index, df in enumerate(dataframes):
                print(f"\n##########################\n\tTable {index}\n##########################\n")
                display(df)
    
        return dataframes
    return []
```

# 2018

For 2018 all necessary tables are in the same PDF so let's extract them all here.

```python
tables_2018 = extract_tables("./files/2019 - BFI yearbook 2019 - 888.pdf", pages="11,34,70")
```

Looks like the first table is the monthly breakdown we're looking for.

<span style="color:red">Mention insertion of `Year` everywhere</span>

```python
admissions_2018 = (
    tables_2018[0]
    .loc[2:13,[0, 2]] # drop first two and last two rows, keep month and 2018 admissions column
)

admissions_2018.columns = ["Month", "Admissions (million)"]

admissions_2018.insert(0, "Year", 2018)

admissions_2018.head()
```

```python
admissions_2018.tail()
```

#### Genres (2018)

```python
genres_2018 = (
    tables_2018[1]
    .loc[4:21] # drop first 4 and last 2 rows
    .drop(columns=[2, 4])
)

genres_2018.insert(0, "Year", 2018)

genres_2018.columns = ["Year", "Genre", "Number of releases", "Gross box office (£ million)", "Top performing title"]

genres_2018.head()
```

```python
genres_2018.tail()
```

#### Distributors (2018)

```python
distributors_2018 = (
    tables_2018[3]
    .loc[5:] # drop first 5 rows
    .drop(index=[15, 17, 18]) # drop rows relating to totals EXCEPT the row totalling "others"
)

distributors_2018.insert(0, "Year", 2018)

distributors_2018.columns = ["Year", "Distributor", "Market share", "Films on release", "Box office gross (£ million)"]

distributors_2018.head()
```

```python
distributors_2018.tail()
```
